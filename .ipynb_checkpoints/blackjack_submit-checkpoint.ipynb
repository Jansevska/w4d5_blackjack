```python
import random

class Card:
    def __init__(self):
        self.card_categories = ['Hearts', 'Diamonds', 'Clubs', 'Spades'] 
        self.cards_list = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King'] 
  
    def get_card_value(self, card): 
        if card[0] in ['Jack', 'Queen', 'King']: 
            return 10
        elif card[0] == 'Ace': 
            return 11
        else: 
            return int(card[0])
        
    def create_deck(self):
        self.deck = [(card, category) for category in self.card_categories for card in self.cards_list]
        random.shuffle(self.deck)
        return self.deck
               
    def player_cards(self):
        new_card = self.deck.pop()
        return new_card
    
    def dealer_cards(self):
        new_card = self.deck.pop()
        return new_card

        
class Game:
    def __init__(self):
        self.player_cards = []
        self.dealer_cards = []
        self.player_score = 0
        self.dealer_score = 0
        self.card = Card()
        self.cards = self.card.create_deck()
        
    def start(self):
        for _ in range(2):
            self.dealer_cards.append(self.cards.pop())
            self.player_cards.append(self.cards.pop())
        print(self.player_cards)
        self.player_turn = True
        
    def player(self):
        new_card = self.card.player_cards() 
        self.player_cards.append(new_card)
        return new_card
        
    def score(self):
        player_score = sum(self.card.get_card_value(card) for card in self.player_cards) 
        dealer_score = sum(self.card.get_card_value(card) for card in self.dealer_cards)
        
        if player_score > 21: 
            print("Dealer Has:", self.dealer_cards) 
            print("Dealer's Score:", dealer_score) 
            print("You Have:", self.player_cards) 
            print("Your Score:", player_score) 
            print("Dealer wins\n(Player Loss Because Player Score is exceeding 21)")
        elif self.player_turn:
            print("You Have:", self.player_cards) 
            print("Your Score:", player_score)
            print("Dealer Has:", self.dealer_cards[1:]) 
        else:
            while dealer_score < 17: 
                new_card = self.cards.pop() 
                self.dealer_cards.append(new_card) 
                dealer_score += self.card.get_card_value(new_card)

            print("Dealer Has:", self.dealer_cards) 
            print("Dealer's Score:", dealer_score) 
            print("\n")
            
            if player_score == 21 and len(self.player_cards) == 2:
                print("________________________________\n")
                print("***     B l a c k j a c k    ***\n")
                print(" C O N G R A T U L A T I O N S ! ")
                print("________________________________")
            elif dealer_score > 21: 
                print("Dealer Has:", self.dealer_cards) 
                print("Dealer's Score:", dealer_score) 
                print("You Have:", self.player_cards) 
                print("Your Score:", player_score) 
                print("You Won!!!\n(Dealer Loss Because Dealer Score is exceeding 21)") 
            elif player_score > dealer_score: 
                print("Dealer Has:", self.dealer_cards) 
                print("Dealer's Score:", dealer_score) 
                print("You Have:", self.player_cards) 
                print("Your Score:", player_score) 
                print("You won!!!\n(Player Has High Score than Dealer)") 
            elif dealer_score > player_score: 
                print("Dealer Has:", self.dealer_cards) 
                print("Dealer's Score:", dealer_score) 
                print("You Have:", self.player_cards) 
                print("Your Score:", player_score) 
                print("Dealer wins\n(Dealer Has High Score than Player)") 
            else: 
                print("Dealer Has:", self.dealer_cards) 
                print("Dealer's Score:", dealer_score) 
                print("You Have:", self.player_cards) 
                print("Your Score:", player_score) 
                print("It's a tie.")
        
    def finish(self):
        self.player_cards = []
        self.dealer_cards = []
        self.player_score = 0
        self.dealer_score = 0
        self.player_turn = True
        print("Bye Bye")
        

def main():
    card = Card()
    card.create_deck()
    game = Game()
    
    print("________________________________________________________________\n")
    print("*  *  *  *  *     B  L  A  C  K  J  A  C  K        *  *  *  *  *")
    print("________________________________________________________________")
    print("\n")
    
    while True:
        choice = input("Enter 'play' to play or 'stop' to stop and quit playing > ")
        if choice == 'stop':
            game.finish()
            break
        elif choice == 'play':
            game.start()
            card.player_cards()
            game.score()
            while True:
                choice = input('Enter "hit" to request another card or "stand" to play with what you have: ').lower()
                if choice == "hit":
                    game.player()
                    game.score()
                    if sum([game.card.get_card_value(card) for card in game.player_cards]) > 21:
                        game.finish()
                        break
                elif choice == "stand":
                    game.player_turn = False
                    card.dealer_cards()
                    game.score()
                    game.finish()
                    break
                else:
                    print("Invalid input. Please enter 'hit' or 'stand' > ")
            

main()           
```
